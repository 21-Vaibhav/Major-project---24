```python
from datasets import load_dataset
from transformers import BertTokenizer

# Load the dataset
ds = load_dataset("santoshtyss/indian_courts_cases")

# Initialize BERT tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Preprocessing function to tokenize the input text
def preprocess_text(examples, tokenizer=tokenizer):
    # Tokenize the text into word-level tokens
    encoding = tokenizer(
        examples['text'],
        padding="longest",
        truncation=True,
        max_length=512,
        return_tensors="pt"
    )
    return encoding

# Apply the tokenizer to the dataset with parallel processing
ds = ds.map(
    preprocess_text,
    batched=True,
    num_proc=4,  # Use 4 processes for faster mapping
    fn_kwargs={'tokenizer': tokenizer}  # Pass tokenizer as a keyword argument
)

# Print sample to check if preprocessing worked
print(ds['train'][0])  # Check a single processed example from the training set
```

```python
import torch
import torch.nn as nn
from transformers import BertModel

class HAN(nn.Module):
    def __init__(self, hidden_size=256, num_classes=2):
        super(HAN, self).__init__()
        # Pre-trained BERT for word embeddings
        self.bert = BertModel.from_pretrained("bert-base-uncased")

        # Sentence-level BiLSTM
        self.sentence_lstm = nn.LSTM(input_size=self.bert.config.hidden_size,
                                      hidden_size=hidden_size,
                                      num_layers=1, bidirectional=True, batch_first=True)

        # Word-level BiLSTM
        self.word_lstm = nn.LSTM(input_size=self.bert.config.hidden_size,
                                  hidden_size=hidden_size,
                                  num_layers=1, bidirectional=True, batch_first=True)

        # Sentence-level attention
        self.attn = nn.Linear(hidden_size * 2, 1)

        # Fully connected layer for final output (summary or classification)
        self.fc = nn.Linear(hidden_size * 2, num_classes)

    def forward(self, input_ids, attention_mask):
        # Word-level embeddings using BERT
        word_embeddings = self.bert(input_ids=input_ids, attention_mask=attention_mask)[0]

        # Word-level BiLSTM
        word_out, _ = self.word_lstm(word_embeddings)

        # Apply word-level attention
        word_attention_weights = torch.softmax(self.attn(word_out), dim=1)
        word_representation = torch.sum(word_attention_weights * word_out, dim=1)

        # Sentence-level processing
        sentence_out, _ = self.sentence_lstm(word_representation.unsqueeze(1))

        # Apply sentence-level attention
        sentence_attention_weights = torch.softmax(self.attn(sentence_out), dim=1)
        sentence_representation = torch.sum(sentence_attention_weights * sentence_out, dim=1)

        # Final fully connected layer
        output = self.fc(sentence_representation)

        return output
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
from transformers import AdamW, get_linear_schedule_with_warmup

# Custom Dataset class
class IndianCourtsDataset(Dataset):
    def __init__(self, dataset):
        self.input_ids = dataset['input_ids']
        self.attention_mask = dataset['attention_mask']

        # Generate labels (modify this logic as needed)
        self.labels = [1 if len(text) > 500 else 0 for text in dataset['text']]

    def __len__(self):
        return len(self.input_ids)

    def __getitem__(self, idx):
        return {
            'input_ids': torch.tensor(self.input_ids[idx], dtype=torch.long),
            'attention_mask': torch.tensor(self.attention_mask[idx], dtype=torch.long),
            'labels': torch.tensor(self.labels[idx], dtype=torch.long)
        }

# HAN Model Implementation
class HierarchicalAttentionNetwork(nn.Module):
    def __init__(self, vocab_size, embedding_dim=100, hidden_size=50, num_classes=2):
        super(HierarchicalAttentionNetwork, self).__init__()

        # Word-level Attention
        self.embedding = nn.Embedding(vocab_size, embedding_dim)
        self.word_gru = nn.GRU(embedding_dim, hidden_size, bidirectional=True, batch_first=True)
        self.word_attention = nn.Linear(hidden_size * 2, 1)

        # Sentence-level Attention
        self.sentence_gru = nn.GRU(hidden_size * 2, hidden_size, bidirectional=True, batch_first=True)
        self.sentence_attention = nn.Linear(hidden_size * 2, 1)

        # Classification layer
        self.fc = nn.Linear(hidden_size * 2, num_classes)

    def forward(self, input_ids, attention_mask):
        # Embedding layer
        embedded = self.embedding(input_ids)

        # Apply attention mask
        mask = attention_mask.unsqueeze(-1).expand(embedded.size())
        embedded = embedded * mask

        # Word-level attention
        word_outputs, _ = self.word_gru(embedded)
        word_attention_weights = torch.softmax(self.word_attention(word_outputs), dim=1)
        word_context = torch.sum(word_attention_weights * word_outputs, dim=1)

        # Sentence-level attention
        sentence_outputs, _ = self.sentence_gru(word_context.unsqueeze(1))
        sentence_attention_weights = torch.softmax(self.sentence_attention(sentence_outputs), dim=1)
        document_vector = torch.sum(sentence_attention_weights * sentence_outputs, dim=1)

        # Classification
        output = self.fc(document_vector)
        return output

# Training function
def train_han_model(ds, model, tokenizer, num_epochs=3, batch_size=8, learning_rate=1e-5):
    # Create custom dataset
    train_dataset = IndianCourtsDataset(ds['train'])

    # Create DataLoader
    train_loader = DataLoader(
        train_dataset,
        batch_size=batch_size,
        shuffle=True
    )

    # Move model to GPU if available
    device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
    model.to(device)

    # Optimizer and Loss
    optimizer = AdamW(model.parameters(), lr=learning_rate)
    loss_fn = nn.CrossEntropyLoss()

    # Learning rate scheduler
    total_steps = len(train_loader) * num_epochs
    scheduler = get_linear_schedule_with_warmup(
        optimizer,
        num_warmup_steps=0,
        num_training_steps=total_steps
    )

    # Training loop
    for epoch in range(num_epochs):
        model.train()
        total_loss = 0

        for batch in train_loader:
            # Move batch to device
            input_ids = batch['input_ids'].to(device)
            attention_mask = batch['attention_mask'].to(device)
            labels = batch['labels'].to(device)

            # Zero gradients
            optimizer.zero_grad()

            # Forward pass
            outputs = model(input_ids, attention_mask)

            # Compute loss
            loss = loss_fn(outputs, labels)

            # Backward pass
            loss.backward()

            # Optimizer step
            optimizer.step()
            scheduler.step()

            total_loss += loss.item()

        # Print epoch statistics
        print(f"Epoch {epoch+1}/{num_epochs}")
        print(f"Average Loss: {total_loss / len(train_loader):.4f}")

    return model

# Main execution
def main():
    # Ensure tokenizer is imported and initialized
    from transformers import BertTokenizer
    tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

    # Determine vocabulary size
    vocab_size = tokenizer.vocab_size

    # Initialize HAN model
    model = HierarchicalAttentionNetwork(
        vocab_size=vocab_size,
        embedding_dim=100,
        hidden_size=50,
        num_classes=2  # Binary classification example
    )

    # Train the model
    trained_model = train_han_model(ds, model, tokenizer)
    torch.save(trained_model, 'han_model_complete.pth')


# Run the main function
if __name__ == '__main__':
    main()
```
